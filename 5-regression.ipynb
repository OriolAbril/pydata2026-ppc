# Posterior predictive checks for regression models

Depending on the type of regression we can use all the pertinent posterior predictive checks we have already seen.
In addition, we can complement them with plots in the x vs y space which can give extra intuition. In ArviZ the main function to this end is `plot_lm`, but you should be able to define custom functions and take advantage of automatic facetting and/or aesthetics mapping if you need a different layout for each subplot

```python
import arviz as az
import numpy as np
import xarray as xr

az.style.use("arviz-variat")
```

```python
dt = az.load_arviz_data("roaches_zinb")
```

```python
az.plot_lm(dt)
```

```python
az.plot_lm(dt, ci_prob=(0.5, 0.75, 0.9));
```
