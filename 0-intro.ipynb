## What is the posterior predictive distribution?
We use posterior predictive to refer to the distribution _a posteriori_ over observations.

$$
p(\tilde{y}) = \int_\Theta p(\tilde{y} | \theta) p(\theta | y) d\theta
$$

That is, we use the observational distribution from which the likelihood was defined and we use it to generate predictions.
The idea behind posterior predictive checks is very general and simple: if a model is good it should be able generate data resembling the observed data.
Consequently, most of what we'll explore today will also apply to any generative model capable of generating predictions probabilistically.

The general algorithm to generate samples from the posterior predictive distribution for Bayesian models is the following:

1. Draw $N$ realizations from the posterior distribution.
2. For each draw, simulate new data using the observational distribution.
