## Rootogram posterior predictive checks

Rootograms are a tool for visually assessing the goodness of fit for count data that was introduced in the 70s. The basic idea is very similar to checking T-statistics for frequency of each possible discrete value. To illustrate this we show the comparison between the frequency of the observation being equal to 20 between posterior predictive and observations:

```python
def freq_obs_20(x):
    return (x==20).sum(axis=0)
    
az.plot_ppc_tstat(dt, t_stat=freq_obs_20);
```

```python
pc = az.plot_dist((dt.posterior_predictive == 20).sum("match").dataset)
az.add_lines(pc, (dt.observed_data == 20).sum().dataset);
```

ArviZ includes `plot_ppc_rootogram` that implements a version of the rootogram adapted to Bayesian models. An important difference with standard rootograms is we set the y scale to be a square root scale instead of taking the square roots of the frequencies (what originally gave it its name). This allows for better visualization of the fit for values with low frequency while keeping the plot straightforward to read.

```python
az.plot_ppc_rootogram(dt);
```
