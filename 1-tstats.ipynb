# Posterior predictive checks through T-statistics

We have generated $N$ realizations of data simulated out of the posterior. An initial first check is computing T-statistics like moments over all observations and compare those point estimates to the distribution we get after doing the same for each of the $N$ realizations.

```python
import arviz as az
az.style.use("arviz-variat")
dt = az.load_arviz_data("radon")
```

```python
dt.observed_data.mean("obs_id")
```

```python
dt.posterior_predictive.mean("obs_id")
```

```python
pc = az.plot_dist(dt.posterior_predictive.mean("obs_id").dataset)
az.add_lines(pc, dt.observed_data.mean("obs_id").dataset);
```

```python
az.plot_ppc_tstat(dt, t_stat="median");
```

```python
az.combine_plots(
    dt,
    [
        (az.plot_ppc_tstat, {"t_stat": "std"}),
        (az.plot_ppc_tstat, {"t_stat": "mad"}),
        (az.plot_ppc_tstat, {"t_stat": "iqr"})
    ],
    group="posterior_predictive",
    expand="row"
);
```
