## Calibration plots

The main tool for posterior predictive checks in binary data are calibration plots. The main intuition behind them is comparing the probability of success in bins/sections defined by having similar success probability in the posterior predictive with the average between all observations at those same indexes.

```python
dict(dt.observed_data.sizes)
```

```python
pp_means = dt.posterior_predictive["vote"].mean(("chain", "draw"))
dt.observed_data["vote"][(pp_means < 0.1)]#.mean()
```

```python
dt.observed_data["vote"][(pp_means > 0.9)]#.mean()
```

```python
dt.observed_data["vote"][(0.4 < pp_means) & (pp_means < 0.5)].mean()
```

ArviZ has a `plot_ppc_pava` function that implements a much more robust version of calibration plots that doesn't rely on binning the data and also allows for computing confidence bands from the posterior predictive.

You'll see the y axis shows "CEP", short for “conditional event probabilities.” A CEP represents the probability that an event occurs, given that the classifier assigned a particular predicted probability. These probabilities are computed using the pool adjacent violators algorithm, also called "PAV-adjusted" which gives the name to the plotting function. That being said, these details are not particularly critical to the interpretation of the plot.

```python
az.plot_ppc_pava(dt);
```

Calibration plots can also be used for discrete data. Their range goes between the minimum and the maximum values in `pp_means`. This also means that binary models with very high confidence around the success probability might show a plot with axis between 0.4 and 0.53 only. In very short ranges the uncertainty around the extremes can be much higher than the range of the central line so it often ends up looking like a flat line instead of diagonal; as long as the line is fully inside the bands this is not an issue at all.
